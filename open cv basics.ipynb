```python


img3 = cv2.imread("pictures/virat 8.jpg")

# Fix color conversion
img = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

figure_size = 30
result = cv2.blur(img, (figure_size, figure_size))

plt.figure(figsize=(11, 6))

plt.subplot(121), plt.imshow(img), plt.title("Original")
plt.xticks([]), plt.yticks([])  # Correct the function names
plt.subplot(122), plt.imshow(result), plt.title("Blurred")
plt.xticks([]), plt.yticks([])  # Correct the function names

plt.show()  # Correct the function name
```

```python
img1=cv2.imread("pictures/Professional pic.jpg")
```

```python
(height,width)=img1.shape[0:2]
zoom=cv2.resize(img1,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom=cv2.resize(img1,(int(width*3),int(height*3)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
m=cv2.getRotationMatrix2D((width/2,height/2),30,1)
rotate=cv2.warpAffine(img1,m,(height,width))
plt.imshow(rotate)
```
